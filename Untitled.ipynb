```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
fpath1 = 'Resources/automation_data_by_state.csv'
fpath2 = 'Resources/occupation_salary.csv'
```

```python
salary_df = pd.read_csv(fpath2)
automation_df = pd.read_csv(fpath1)
```

```python
salary_df.rename(columns = {'OCC_CODE':'SOC','ANNUAL':'ANNUAL_only',
    'HOURLY':'HOURLY_only'}, inplace = True)
```

```python
automation_df.rename(columns = {
    'Probability':'Automation_Probability',
    'District of Columbia':'District_of_Columbia',
    'New Hampshire':'New_Hampshire',
    'New Jersey':'New_Jersey',
    'New Mexico':'New_Mexico',
    'New York':'New_York',
    'North Carolina':'North_Carolina',
    'North Dakota':'North_Dakota',
    'Rhode Island':'Rhode_Island',
    'South Carolina':'South_Carolina',
    'South Dakota':'South_Dakota',
    'West Virginia':'West_Virginia',
    'Occupation':'OCCUPATION'
},inplace=True)
```

```python
salary_df["ANNUAL_only"].fillna("0", inplace = True)
salary_df["HOURLY_only"].fillna("0", inplace = True)
salary_df["ANNUAL_only"].replace(True, 1, inplace=True)
salary_df["HOURLY_only"].replace(True, 1, inplace = True)
salary_df.head(10)
```

```python
salary_df["ANNUAL_only"] = salary_df["ANNUAL_only"].astype(int).astype(bool)
salary_df["HOURLY_only"] = salary_df["HOURLY_only"].astype(int).astype(bool)
salary_df.head(10)
```

```python
salary_df['A_MEAN'] = salary_df['A_MEAN'].str.replace(',','')
salary_df['TOT_EMP'] = salary_df['TOT_EMP'].str.replace(',','')
salary_df['A_PCT10'] = salary_df['A_PCT10'].str.replace(',','')
salary_df['A_PCT25'] = salary_df['A_PCT25'].str.replace(',','')
salary_df['A_MEDIAN'] = salary_df['A_MEDIAN'].str.replace(',','')
salary_df['A_PCT75'] = salary_df['A_PCT75'].str.replace(',','')
salary_df['A_PCT90'] = salary_df['A_PCT90'].str.replace(',','')
salary_df.info()
```

```python
salary_df.replace('#', np.nan, inplace=True)
salary_df.head()
```
